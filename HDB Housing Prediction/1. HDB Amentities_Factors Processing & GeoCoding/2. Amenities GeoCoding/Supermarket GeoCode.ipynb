```python
import requests as requests
import pandas as pd
import json
import folium
from folium.plugins import MarkerCluster
```

```python
df = pd.read_excel('Supermarkets.xlsx')
```

```python
df.head()
```

```python
df.info()
```

```python
df["Address"] =  "Singapore " + df["Postcode"].astype(str)
```

```python
df.head()
```

```python
df.info()
```

```python
geo_s ='https://maps.googleapis.com/maps/api/geocode/json'
df["lat"] = ""
df["lng"] = ""


for i in range(len(df)):
    param = {'address': df["Address"][i], 'key': ''}

    response = requests.get(geo_s, params=param)

    json_dict = response.json()
    
    df["lat"][i] = json_dict['results'][0]['geometry']['location']['lat']
    df["lng"][i] = json_dict['results'][0]['geometry']['location']['lng']
```

```python
df.head()
```

```python
df.to_csv("Preprocessed Supermarket.csv")
```
