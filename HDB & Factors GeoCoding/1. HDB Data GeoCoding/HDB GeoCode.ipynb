```python
df["Address"] =  "Singapore " + df.street_name.astype(str) + ' Block ' + df.block.astype(str)
```

```python
df.to_csv("Final.csv")
```

```python
df= df.drop_duplicates('Address', keep='last')
df = df.reset_index(drop=True)
```

```python
df.info()
```

```python
geo_s ='https://maps.googleapis.com/maps/api/geocode/json'
df["lat"] = ""
df["lng"] = ""


for i in range(len(df)):
    param = {'address': df["Address"][i], 'key': ''}

    response = requests.get(geo_s, params=param)

    json_dict = response.json()
    
    print(df["Address"][i])
    print(json_dict['results'][0]['geometry']['location']['lat'])
    print(json_dict['results'][0]['geometry']['location']['lng'])
    df["lat"][i] = json_dict['results'][0]['geometry']['location']['lat']
    df["lng"][i] = json_dict['results'][0]['geometry']['location']['lng']
```
